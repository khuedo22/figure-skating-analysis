```python
dfs = []
for index, season in enumerate(seasons):
    countries_of_competitions = competitions_per_season[index]
    results_for_season = result_links[index]
    for index_of_season, country_of_competition in enumerate(countries_of_competitions):
        results_for_competition = results_for_season[index_of_season]
        for index_of_discipline, url in enumerate(results_for_competition):
            page = requests.get(url)
            soup = BeautifulSoup(page.content, 'html.parser')
            table = soup.find("div", class_='scroll')
            odd_placements = table.find_all('tr', class_='o')
            even_placements = table.find_all('tr', class_='')
            all_placements = odd_placements + even_placements
            skaters = []
            countries = []
            scores = []
            for skater_information in all_placements:
                if skater_information.find('td', class_ = 'c', width = '5%') == None:
                    continue
                skater_name = skater_information.find('a')
                country = skater_information.find('td', class_='c', width='5%')
                score = skater_information.find_all('td', class_='r', width='8%')[2]

                skaters.append(skater_name.text.strip())
                countries.append(flag_to_country(country.text.strip()))
                scores.append(score.text.strip())
            df = pd.DataFrame({'Skater':skaters,
                               'Country Represented':countries,
                               'Score':scores})
            df['Country of Competition'] = country_of_competition
            df['Discipline'] = disciplines[index_of_discipline]
            df['Season'] = season
            dfs.append(df)

figure_skating_df = pd.concat(dfs)   
```

```python
figure_skating_df['Home Competition?'] = figure_skating_df['Country Represented'] == figure_skating_df['Country of Competition']
```

```python
skater_groups = figure_skating_df.groupby(['Skater', 'Season'], as_index=False)
competes_at_home = []
for name, df in skater_groups:
    if (len(df) == 2) & (True in df['Home Competition?'].values):
        competes_at_home.append(df)
```

```python
filtered_df = pd.concat(competes_at_home)
filtered_df = filtered_df.reset_index(drop=True)
```

```python
filtered_df.to_csv('grand_prix_dataset.csv', index=False)
```
